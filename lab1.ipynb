|Variable|Definition|Key         
|:---:|:---:|---: 
|DataTime|дата и время|

|Holiday|выходной день| 0 = No 1 = Yes
|WorkingDay|рабочий день| 0 = No 1 = Yes
|Weather|погода|1 = A   2 = B   3 = C   4 = D
|Temp|температура в Цельсиях|
|Atemp|температура "по ощущениям"|
|Humidity|относительная влажность|
|WindSpeed|скорость ветра|
|Casual|количество инициированных прокатов незарегистрированных пользователей|
|Registered|количество зарегистрированных арендованных пользователей инициировано|
|Count|количество всех арендных плат|

```python
train['temp'].mean() # среднее значение температуры
```

```python
def meanTemp(t):
    return t > 19
train['temp'].map(meanTemp)
train['good_temp'] = train['temp'].map(lambda t: t > 19)
train.head()
```

```python
df_season = pd.DataFrame( #новый dataframe, который расшифровывает названия времён года
    [
        [1, 'spring'],
        [2, 'summer'],
        [3, 'autumn'],
        [4, 'winter']
    ],
    columns=['id_season', 'name_season']
)
df_season
```

```python
pd.merge( #соединим два dataframe
    train,
    df_season,
    left_on='season',
    right_on='id_season'
)
```
